What to: You have been provided with a dataset that lists Ohio Stateâ€™s leading causes of death from the year 2012.

Using the two data points:  Cause of deaths and Percentile. Draw a pie chart to visualize the dataset

#### 1: List the causes of death and percentile in Ohio state in the year 2012

```python
#import required libraries
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline

import pandas as pd
import seaborn as sns
```

```python
#cause of deaths
cause_of_deaths = ['Chronic Disease', 'Unintetional Injuries', 'Alzheimers', 'Infuenza and Pneumonia', 'Sepsis', 'Others']

#percentile
percentile = [62, 5, 4, 2, 1 , 26]
```

#### 2: Draw a pie chart using the two data points

```python
#set the pie chart plot properties
data = percentile
labels = cause_of_deaths

#set figure size
plt.figure(figsize=(10,10))

#explode the largest pie in the data set
explode = (0.05, 0.05, 0.05, 0.05, 0.05, 0.05)

#set pie chart properties
plt.pie(data, labels=labels, explode=explode, autopct='%1.1f%%', startangle=70)

#set axis equal to draw pie as circle
plt.axis('equal')

#set title of the pie chart
plt.title('Analysing Cause of Death')

# show the pie chart
plt.show()
```
