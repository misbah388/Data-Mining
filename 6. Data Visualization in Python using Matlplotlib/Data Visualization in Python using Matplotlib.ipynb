# Step - 1. Import Required Libraries

```python
# import numpy for random numbers generators
import numpy as np

# import matplotlib library
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

# Step 2. Define or Import the dataset

```python
# Generating 15 random numbers in range of 1 to 50
dataset = np.random.randint(1, 50, 15)
dataset
```

# Step 3. Set the plot parameter

```python
# STEP 3 AND 4 are done in the same cell, there is no need of defining a separate function
# here it's in different cells for steps seperation

def plot_parameter():
    # select the style of plot
    style.use('ggplot')
    
    # setting the plot figure size
    plt.figure(figsize=(10,5))

    # plot the random numbers
    plt.plot(dataset, color='b', label='line_one', linestyle='--', linewidth = 2, marker='+')

    # defining the x-axis
    plt.xlabel('range')

    # defining the y-axis
    plt.ylabel('numbers')

    # setting the title
    plt.title('First plot')
```

# Step 4. Display the created plot

```python
# calling the parameter function
plot_parameter()

# applying legends
plt.legend()

# plotting the graph
plt.show()
```

# Hisytogram and Scatter Plot

```python
# import boston dataset from sklearn library
from sklearn.datasets import load_boston
```

```python
# import matplotlib
# already imported at the beginning of the notebook
```

```python
# load boston dataset
dataset = load_boston()
```

```python
# view boston dataset
dataset
```

```python
# define x-axis for the data
x_axis = dataset.data
x_axis
```

```python
# define y-axis for the target
y_axis = dataset.target
y_axis
```

```python
# plot histogram

plt.figure(figsize=(9,9))
style.use('ggplot')

plt.hist(y_axis, bins=50)
plt.title('Histogram')
plt.xlabel('Price in 1000 USD $')
plt.ylabel('No. of Houses')
plt.show()
```

```python
# plot scatter plot


plt.figure(figsize=(9,9))
style.use('ggplot')

plt.scatter(x_axis[:,5], y_axis)
plt.title('Histogram')
plt.ylabel('Price in 1000 USD $')
plt.xlabel('No. of Houses')
plt.show()
```
